# Calibration Curves by Cell Type

This notebook reproduces the built-in `corresponding_Ca_value` formulas from `src/general/RatioToConcentrationConverter.py`, plots all curves together, and saves the figure as `Calibrationscurve.png`.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def jurkat_ca_from_ratio(ratio_value):
    a = 15.48
    b = 1.9
    return a * np.exp(b * ratio_value)


def primary_ca_from_ratio(ratio_value):
    a = 44.05
    b = 1.627
    c = -88.88
    d = -1.311
    return a * np.exp(b * ratio_value) + c * np.exp(d * ratio_value)


def nk_ca_from_ratio(ratio_value):
    a = 8.613 * 10 ** (-9)
    b = 6.735
    c = 26.34
    d = 1.166
    return a * np.exp(b * ratio_value) + c * np.exp(d * ratio_value)


def nk_human_ca_from_ratio(ratio_value):
    a = 50.39
    b = 1.7
    return a * np.exp(b * ratio_value)
```

```python
ratio_values = np.linspace(0.0, 2.2, 500)

curves = {
    "jurkat": jurkat_ca_from_ratio(ratio_values),
    "primary": primary_ca_from_ratio(ratio_values),
    "NK": nk_ca_from_ratio(ratio_values),
    "NK_human": nk_human_ca_from_ratio(ratio_values),
}
```

```python
output_path = "Calibrationscurve.png"

color_map = {
    "jurkat": "tab:blue",
    "primary": "tab:orange",
    "NK": "tab:green",
    "NK_human": "tab:red",
}

plt.figure(figsize=(9, 6))
for label, values in curves.items():
    plt.plot(ratio_values, values, label=label, color=color_map[label], linewidth=2)

plt.xlabel("Ratio value")
plt.ylabel("Corresponding Ca value (nM)")
plt.title("Built-in DARTS Calibration Curves by Cell Type")
plt.grid(True, alpha=0.3)
plt.legend(title="Cell type")
plt.tight_layout()
plt.savefig(output_path, dpi=300, bbox_inches="tight")
plt.show()
```
