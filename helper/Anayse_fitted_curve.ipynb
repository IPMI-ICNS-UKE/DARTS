```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 31.140785029573205
b = 2.0622644818601232
c = -43.28837537323233
d = -9.30649530756311
```

```python
#vorwärtsfunction
def ca_from_ratio(r, a, b, c, d):
    return a * np.exp(b * r) + c * np.exp(d * r)
```

```python
ratio_vals = np.linspace(0, 2.2, 500)  # typischer Bereich deiner Daten
ca_vals = ca_from_ratio(ratio_vals, a, b, c, d)
```

```python
plt.figure(figsize=(7,5))
plt.plot(ratio_vals, ca_vals)
plt.xlabel("Ratio (Fluo4 / FuraRed)")
plt.ylabel("Ca²⁺ concentration (nM)")
plt.title("Calcium Calibration Curve")
plt.grid(True)
plt.show()
```
